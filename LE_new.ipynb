```python
car_data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)
#one hot encoding
car_data = pd.get_dummies(car_data, columns=['Seller_Type', 'Transmission'], drop_first=True)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(car_data.corr(), annot=True)
plt.title('Correlation between the columns')
plt.show()#lighther high corelation and darker low coreleation and we can remove the colms of low corelation but in our case we have only few cols
```

```python
X = car_data.drop(['Car_Name','Selling_Price'], axis=1)
y = car_data['Selling_Price']
```

```python
print("Shape of X is: ",X.shape)
print("Shape of y is: ", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print("X_test shape:", X_test.shape)
print("X_train shape:", X_train.shape)
print("y_test shape: ", y_test.shape)
print("y_train shape:", y_train.shape)
```

```python
scaler = StandardScaler()#to normalize the data removes the mean scales each unit to variance
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("MAE: ", (metrics.mean_absolute_error(pred, y_test)))
print("MSE: ", (metrics.mean_squared_error(pred, y_test)))
print("R2 score: ", (metrics.r2_score(pred, y_test)))#more the r2 score shows how best fit the traing eg to the modek
```

```python
sns.regplot(x=pred, y=y_test)
plt.xlabel("Predicted Price")
plt.ylabel('Actual Price')
plt.title("ACtual vs predicted price")
plt.show()
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, lw=0.6, color="blue", label="original")
plt.plot(x_ax, pred, lw=0.8, color="red", label="predicted")
plt.title("target test and predicted data")
plt.legend()
plt.show()
```
